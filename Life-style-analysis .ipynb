## Feature engineering. 

An important aspect of building a reliable and machine learning model is processing data to a form the model will understand and be less bias towards. Feature engineering is the process taking a raw data, extracting, and organising important feature from it, to fit model operation method (Sreemany, 2021). ## Target Encoding
To choose the right encoding method for our target variable, it is important to understand what each unique variable mean. Encoding method works in different ways, for instance, encoding a nominal categorical variable is different from encoding an ordinal categorical variable. According to data dictionary our target variable i 'Age_Group'. This contains two unique age groups namely 'Adult' and 'Senior'. The 'adult' are respondents that are below the age 65years old and 'seniors' are respondents above 65years old. This in context to my understanding means senior respondents and more aged than adult respondents. 
## How does this impact encoding method?
Choosing the encoding method to use is based on the relationship the categories have with each other and the context of task. In our target variable, two fair and relevant argument can be made, the first is that there is no order between senior and adult respondent i.e. we may say 'senior' respondents are independent of 'adult'. And we can also fairly ague than a person above the age of 65 is ranked older than a person under the age 65. The question now is "is it important that our encoding method reveal the ranking or not?" I believe the fair answer would be "it depends", bearing in mind no single method works for all cases, the context on the task at hand would be a determining factor of what our encoding method will capture.  
### Types of encoding method
There is a different types of encoding methods but i will evaluate 2 based on the task at hand.
1. Label Encoder: This encoder reflects sequence or order. It is used for categorical variable that are ordinal (Sreemany, 2021). (Encoding categorical data note, 2023) encourages to use label encoder when categorical target variable needs to be encoded for a classification predictive modelling. In Label encoding, each label is converted into an integer value (Sreemany, 2023).
2. One Hot encoding: It is used when categorical variables are nominal. For each label/category in categorical variable, one hot encoder I will create a new feature, and represent the presence of the label with 1 and the absence of the label as 0 in each newly created feature (Sreemany, 2021).

#### Why chose Label Encoder
Based on the definitions and context of this age-group prediction analysis, i choose to use label encoding because model in this analysis is built to ultimately predict the age-group of a respondent based on their age and live styles for example, it is important for machine learning to be able to differentiate between a 75-year-old and 25 year old respondent. One Hot encoder I believe is not suited for this task because there is ranking/ order in the age-group, because it is not equal if a person is a senior or an adult respondent. One Hot encoder will also create sparsity in the data, because lots of 0's are created hence creating more noise/complexity for model (Sreemany, 2021).

## Curse of dimensionality
Dimensionality of a dataset is the number of features/ variables in a dataset. Curse of dimensionality is a challenge because of high dimensionality in a data, as the number of dimensions in a dataset increase, so does the number of observations required to make meaningful conclusion about the dataset. (The cure of Dimensionality, 2023) whiles machine learning models can handle finding pattern and relationships in a highly dimensional dataset, humans are largely limited in this regards. This is why need arise for dimensionality reduction technique.

## Explain dimensionality reduction techniques.
These are techniques that are used to reduce dimension the name implies. Dimensionality reduction technique generally focuses on finding the optimal number of features needed to preserve most of the variance in a dataset. This principle is based on 'HUGHES PHENOMENON'. As shown in fig.4, as the number of features increases, the classifier’s performance increases as well until we reach the optimal number of features, adding more features after this point without increasing the training set causes a downward performance of the classifier.

Fig.4

## PCA VS LDA
Principal component analysis is an unsupervised dimensionality reduction technique, it is an algorithm that finds the optimal number of features needed to explain majority of the variance in a highly dimensional dataset.  PCA is fundamentally a dimensionality reduction algorithm, but it can also a visualization tool, for noise filtering, for 
feature extraction and engineering (PCA, 2023). PCA centres the data around its mean and find the eigenvector which represent the direction of maximum variance and eigenvalues which represent the amount of explained variance to determine the optimal number of variables that has the most explained variance (Vungarala, 2023). 

### LDA 
Linear discriminate analysis is commonly used for classification problem. It is a supervised dimensionality reduction technique that aims to find the linear combination of features that best classify classes in a dataset (Vungarala, 2023). LDA does this through the following process: (LDA, 2023)
1. Calculate the mean and covariance matrix for each class of data
2. calculate the between and within class scatter matrix. Within-class matrix is the measurement of how individual data points within the same class vary from their class mean WHILE the between-class matrix measures how different classes are separated from each other. 
3. compute the eigenvector and eigenvalues.
### Comparison
In comparison, they are both robust dimensionality reduction technique their mode of operation however differs. pca aims to find the directions of maximum variance in the data, while LDA is a supervised method that aims to find the projection that best separates the classes in the data (Vungarala, 2023).

# Hyperparameter tunning

# Hyper-parameters

The following hyperparameters are present in a random forest classifier. Note that most of these hypereparameters are actually of the decision trees that are in the forest.


- **n_estimators**: integer, optional (default=10): The number of trees in the forest.
- **criterion**: string, optional (default=”gini”)The function to measure the quality of a split. Supported criteria are “gini” for the Gini impurity and “entropy” for the information gain. Note: this parameter is tree-specific.
- **max_features** : int, float, string or None, optional (default=”auto”)The number of features to consider when looking for the best split:
    - If int, then consider max_features features at each split.
    - If float, then max_features is a percentage and int(max_features * n_features) features are considered at each split.
    - If “auto”, then max_features=sqrt(n_features).
    - If “sqrt”, then max_features=sqrt(n_features) (same as “auto”).
    - If “log2”, then max_features=log2(n_features).
    - If None, then max_features=n_features.
    - Note: the search for a split does not stop until at least one valid partition of the node samples is found, even if it requires to effectively inspect more than max_features features.
- **max_depth** : integer or None, optional (default=None)The maximum depth of the tree. If None, then nodes are expanded until all leaves are pure or until all leaves contain less than min_samples_split samples.
